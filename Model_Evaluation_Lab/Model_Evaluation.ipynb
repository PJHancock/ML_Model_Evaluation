#### Step 1: Read in and appropriately process the training data from CS270Train.csv

```python
import pandas as pd
from sklearn.model_selection import train_test_split

CS270Train = pd.read_csv("CS270Train.csv")
CS270Train_df = pd.DataFrame(CS270Train)
X = CS270Train_df.drop(columns=["class", "id"])
y = CS270Train_df["class"]

# Creating a test/val split to make comparisons between the models
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = .2, random_state = 42)

display(X_train)
display(y_train)
```

#### Step 2: Create the first model and test it

```python
# Initialize and test KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

# Create function to test model and save best
def test_model(model, best_model, best_model_val_acc):
    model.fit(X_train, y_train)
    val_acc = model.score(X_val, y_val)
    if val_acc > best_model_val_acc:
        best_model = model
        best_model_val_acc = val_acc
    return best_model, best_model_val_acc

# Test KNN models with n_neighbors 1-15 and save the best one
best_KNN_model = None
best_KNN_model_val_acc = 0
for n in range(1,15):
    KNNmodel = KNeighborsClassifier(n_neighbors=n)
    best_KNN_model, best_KNN_model_val_acc = test_model(KNNmodel, best_KNN_model, best_KNN_model_val_acc)

# Print out results of best KNN model
print(f"Train accuracy for best KNN model: {best_KNN_model.score(X_train, y_train)}")
print(f"Val accuracy for best KNN model: {best_KNN_model_val_acc}")
```

#### Step 3: Create the second model and test it

```python
# Initialize and test Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Use test_model function above to find best DTCModel
best_DTC_model = None
best_DTC_model_val_acc = 0
for max_depth in range(2, 20):
    DTCmodel = DecisionTreeClassifier(max_depth=max_depth, random_state = 42)
    best_DTC_model, best_DTC_model_val_acc = test_model(DTCmodel, best_DTC_model, best_DTC_model_val_acc)

# Print out results of best DTC Model
print(f"Train accuracy for best DTC model: {best_DTC_model.score(X_train, y_train)}")
print(f"Val accuracy for best DTC model: {best_DTC_model_val_acc}")
```

#### Step 4: Create the third model and test it

```python
# Initialize and test Logistic Regression
from sklearn.linear_model import LogisticRegression

# Use test_model function above to find best Logistic Regression Model
best_LR_model = None
best_LR_model_val_acc = 0
for penalty in ['l2', None]:
    LRModel = LogisticRegression(penalty=penalty)
    best_LR_model, best_LR_model_val_acc = test_model(LRModel, best_LR_model, best_LR_model_val_acc)

# Print out results of best Logisitic Regression Model
print(f"Train accuracy for best LR model: {best_LR_model.score(X_train, y_train)}")
print(f"Validation accuracy for best LR model: {best_LR_model_val_acc}")
```

```python
# Extra classifier test

# Initialize and test Naive Bayes Model
from sklearn.naive_bayes import GaussianNB

# Use test_model function above to find best GaussianNB Model
GNBModel = GaussianNB().fit(X_train,y_train)
print(f"Train accuracy for GNB model: {GNBModel.score(X_train,y_train)}")
print(f"Validation accuracy for GNB model: {GNBModel.score(X_val, y_val)}")
```

#### Step 5: Select the best model.

```python
print(f"Validation accuracy of best KNeighborsClassifier model: {best_KNN_model.score(X_val, y_val)}")
print(f"Validation accuracy of best DecisionTreeClassifer model: {best_DTC_model.score(X_val, y_val)}")
print(f"Validation accuracy of best LogisticRegression model: {best_LR_model.score(X_val, y_val)}")
print(f"validation accuracy of best Gaussian Naive Bayes model: {GNBModel.score(X_val, y_val)}")

# Save the KNN model as the best model
best_model = best_KNN_model
```

#### Step 5a: Report the best model

**By creating a train/val split with the train data, I was able to find the model that generalized the data the best which should perform the best when new data is introduced, i.e. the test data. This means that the model was able to get a high train accuracy while not being overfit to the train data. For each model, I also iterated over important hyperparameters such as n_neighbors for KNeighborsClassifier, max_depth for DecisionTreeClassifier, and penalty for LogisticRegression. This allowed me to find the optimal or near optimal model of that type for the data being tested. Finally, I tested the best model of each model type on the validation data to determine which model performs the best overall.  As seen above, the best performing model is the KNeighborsClassifier model with a validation accuracy of 0.84 while the DecisionTreeClassifier model and LogisiticRegression model only achieve a validation accuracy of 0.77 and the Gaussian Naive Bayes model with a validation accuracy of 0.75.**

#### Step 6: Run the best model on the test data in CS270Test.csv and report the results

```python
from sklearn.metrics import accuracy_score
# Load the test data
CS270Test = pd.read_csv("CS270Test.csv")
CS270Test_df = pd.DataFrame(CS270Test)

# Separate into X and y
X_test = CS270Test_df.drop(columns=["class","id"])
y_test = CS270Test_df["class"]

predictions = best_model.predict(X_test)
print(f"Accuracy on test data using the best Decision Tree Classifier Model: {accuracy_score(y_test, predictions)}")
```
